```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load your dataset (replace with your file path)
df = pd.read_csv("C:/Users/luvpa/Downloads/cleaned_merged_heart_dataset.csv")  # Replace with your CSV file

# Step 3: Set the visual style for plots
sns.set(style="whitegrid")

# Step 4: List of numerical features to analyze
numerical_features = ['age', 'trestbps', 'chol', 'thalachh', 'oldpeak']

# Step 5: Bivariate analysis: Numerical vs. Target
for feature in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='target', y=feature, data=df)  # Removed palette to avoid warning
    plt.title(f'Boxplot of {feature} by Target', fontsize=14)
    plt.xlabel('Target', fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.show()

# Step 6: List of categorical features
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Step 7: Bivariate analysis: Categorical vs. Target
for feature in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=feature, hue='target', data=df, palette='Set1')  # hue='target' used
    plt.title(f'{feature} Distribution by Target', fontsize=14)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.legend(title='Target', loc='upper right')
    plt.show()
```
